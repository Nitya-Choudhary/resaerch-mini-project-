```python
df = pd.read_csv("Coffe_sales.csv")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
df['money'].hist(bins=30, color="peru", edgecolor="black")
plt.title("Distribution of Sale Amounts")
plt.xlabel("Sale Amount (money)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,5))
hourly_sales = df.groupby("hour_of_day")["money"].sum()
hourly_sales.plot(kind='line', marker='o', color="darkred")
plt.title("Total Sales by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```
